# 1. Problem Statement

**My objective for this project is to build and iteratively optimize a linear regression model to predict the price of houses on sale in Ames, Iowa.** <br/> 

My definition for an optimal model score here is to achieve a score which beats the null score and minimizes the difference between training and testing scorea so that the model is tuned for a good balance betwen bias and variance. I find this to be the most important metric on which to evaluate the model because I want to increase the generalizability of the model's predictive power. <br/>


![house_sold_image.png](../images/house_sold_image.png)





**THE DATA**  

The [dataset](https://git.generalassemb.ly/DSI-US-7/project_2/tree/master/datasets) is sourced from the Ames Assessor's Offoice and contains 82 features (including the target feature) and 2930 observations. Throguhough this project, I've frequently referred to the [description file](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt) which includes information on all the variables and variable types included in the dataset. 


**A COMMENT ON MY METHODOLOGY**

Beyond guessing simple things like the fact that square footage and location should have some correlation with house prices, I don't know much about what variables most impactfully drive house prices. Due to lack of domain expertiese and time restrictions that prevent me to consult subject matter experts or to do outside research on my own, I choose to approach this problem in a very data driven way. More specifically, beyond jsut a very few feature engineering tricks based on what my common sense tells me should prove useful for the model, I will have as little to do as possible when it comes to choosing the features. I will instead feed close to most of the features into a pipeline and utilize scikit-learn transformers such k_select_best and other techniques like regularization to automatically reduce the number of features for my model to the most relevant subset. 

Since this approach implements a few transformations, my model is going to lose its interpretability. Therefore, the model is not tuned for explanability, but it is rather tuned for predictability. 

```python
# import all the libraries that will be used throughout this notebook 

import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, PowerTransformer
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score

%matplotlib inline
```

# 2. Data Cleaning and Transformation 

## Importing the Data 

```python
# import the data
train = pd.read_csv('./datasets/train.csv')
test = pd.read_csv('./datasets/test.csv')

# engineer a feature for the total square footage of the house 
train['sf'] = train['Total Bsmt SF'] + train['Gr Liv Area']
test['sf'] = test['Total Bsmt SF'] + test['Gr Liv Area']

# remove houses with larger than 4000 sq feet (as advised in the variables description file)
train.drop(index = list(train[train['sf'] > 4000].index), inplace = True)

train.rename(columns = {'SalePrice': 'target'}, inplace = True) 
#train.drop(columns = 'SalePrice', inplace = True)
```

**Here I will concatenate the train and the test data because in order to submit my results to Kaggle, I need to run my model on my test data which must undergo every featrue transformation in the same way as does my train data.**

```python
# create a column which will indicate for every row which data set it originally comes from. This will help later to split the dataframe back to the original train and test sets after all data cleaning and features transformation is done. 

train['set'] = 'train'
test['set'] = 'test'

# concatenate train and test data  
df = pd.concat([train, test], axis = 0)
df.shape
```

## Handling the Missing Values 

```python
plt.figure(figsize = (20,6))
missing_vals = pd.DataFrame(df.isnull().sum())
missing_vals['cols'] = missing_vals.index
missing_vals = missing_vals[missing_vals[0]>0]
missing_vals = missing_vals.sort_values(by = 0, ascending = False)
plt.bar(missing_vals['cols'], missing_vals[0])
plt.xticks(rotation = 90);
```

#### The missing values will need to be handled differently for features of different types (categorical (ordinal, nominal) and numeric). So here I create variables for numeric and categorical features so that each type can be accessed and handled separately. 

```python
# drop the ID and PID columns as they bring no new information to the dataset given that each row has a unique index 
df.drop(columns = ['Id', 'PID','Pool QC', 'Pool Area', 'Misc Feature', 'Misc Val', 'Alley', 'Fence'], axis = 1, inplace = True)
```

```python
# create a variable for numeric features 
num = [column for column in df.columns if df[column].dtypes == float]
num.remove('Bsmt Full Bath')
num.remove('Bsmt Half Bath')
num.remove('Garage Cars')
num.remove('Garage Yr Blt')

num.extend(['Lot Area', '1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF', 'Gr Liv Area', 'Garage Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch'])
```

```python
print(list(num))
```

#### Fill the NA's with feature mean for any given feature for numeric features

```python
# num.remove('Overall Cond')
# num.remove('Overall Qual')
# num.remove('Year Built')
# num.remove('Year Remod/Add')
# num.remove('Yr Sold')

for i in num:
    df[i] = df[i].fillna(value = df[i].mean())
    
# df[num].isnull().sum()
```

```python
# create a variable for nominal features 
nom = ['MS SubClass','MS Zoning', 'Street','Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 'Roof Style', 'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 'Foundation', 'Heating', 'Central Air', 'Electrical', 'Garage Type', 'Sale Type', 'Land Contour', 'Lot Config', 'Neighborhood',]

nom = list(set(nom))

# create a variable for ordinal features 
ordin = ['Bedroom AbvGr', 'Mo Sold', 'Fireplaces',
 'Full Bath', 'Half Bath','Bsmt Full Bath', 'Bsmt Half Bath',
 'Kitchen AbvGr', 'Mo Sold','Overall Cond',
 'Overall Qual','TotRms AbvGrd',
 'Year Remod/Add',
  'Lot Shape', 'Utilities', 'Land Slope','Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Heating QC', 'Kitchen AbvGr', 'Kitchen Qual', 'TotRms AbvGrd', 'Functional', 'Fireplaces', 'Fireplace Qu', 'Garage Finish',  'Garage Cars', 'Garage Cond', 'Paved Drive', 'Mo Sold',  'Garage Qual']
```

```python
# num.extend(nom)
# num.extend(ordin)
# set(num) - set(list(df.columns))
```

#### Fill NA's with 'none' for categorical features (both ordinals and nominals) because in the case of the categorical features, NA's are not really missing data, but simply an indication that a given observation of a house does not have the item the feature represents.

```python
for i in nom:
    df[i].fillna(value = 'none', inplace = True)
```

```python
for i in ordin: 
    df[i].fillna(value = 'none', inplace = True)
```

```python
# create a variable for features that represent time to easily access them if/when necessary
time = ['Year Built','Garage Yr Blt','Year Remod/Add', 'Yr Sold']
```

## Transform the Data via Label Encoding and Dummy Variables 

### Label Encoding of Ordinal Features

#### Ordinal features are categorical variables that have some order among their levels. All ordinal features need to be label encoded into numeric values that preserve the order of the feature. For some of the ordinal features, all levels are label encoded, while for others, some labels are binned and mapped to the same numeric value to decrease the number of labels and therefore, the gradnularity of the data overall. 

```python
# label encoding of ordinal features 

df['Functional'].replace({'Typ':5, 'Mod':4, 'Min2':3, 'Maj1':2, 'Min1':3, 'Sev':1, 'Sal':1, 'Maj2':2}, inplace = True )

df['Bsmt Exposure'].replace({'No':1, 'Gd':4, 'Av':3, 'none':0, 'Mn':2}, inplace = True)

df['Garage Qual'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['Paved Drive'].replace({'Y':2, 'N':0, 'P':1}, inplace = True)

df['Exter Qual'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['Land Slope'].replace({'Gtl':1, 'Sev':3, 'Mod':2}, inplace = True)

df['Exter Cond'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['Garage Cond'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['TotRms AbvGrd'].replace({6:2,  8:3,  5:2,  7:3,  9:3,  4:2, 11:4, 12:4, 10:4,  3:1,  2:1, 
13:4}, inplace = True)

df['Garage Finish'].replace({'RFn':1, 'Unf':1, 'Fin':2, 'none':0}, inplace = True)

df['Heating QC'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['BsmtFin Type 1'].replace({'GLQ':3, 'Unf':1, 'ALQ':3, 'Rec':2, 'none':0, 'BLQ':2, 'LwQ':1}, inplace = True)

df['BsmtFin Type 2'].replace({'GLQ':3, 'Unf':1, 'ALQ':3, 'Rec':2, 'none':0, 'BLQ':2, 'LwQ':1}, inplace = True)

df['Kitchen Qual'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1}, inplace = True)

df['Bsmt Qual'] = df['Bsmt Qual'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1})

df['Fireplace Qu'] = df['Fireplace Qu'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1})

df['Utilities'] = df['Utilities'].replace({'AllPub': 4, 'NoSeWa': 2, 'NoSewr': 3})

df['Bsmt Cond'] = df['Bsmt Cond'].replace({'TA':2, 'Gd':3, 'Fa':1, 'none':0, 'Ex':3, 'Po':1})

df['Lot Shape'] = df['Lot Shape'].replace({'IR1':1, 'Reg':2, 'IR2':3, 'IR3':4})

df['Overall Cond'] = df['Overall Cond'].replace({8:3, 5:2, 7:3, 6:2, 3:1, 9:3, 2:1, 4:2, 1:1})

df['Overall Qual'] = df['Overall Qual'].replace({8:3, 5:2, 7:3, 6:2, 3:1, 9:3, 2:1, 4:2, 1:1, 10:3})
```

```python
# for some ordinals, some 'none's remain. Let's clean that up...
for i in ordin:
    if 'none' in df[i]:
        df[i].replace({'none': 0}, inplace = True)
```

```python
# I like to save data as I make major transformations to it. Here I save all ordinal and numeric data into a separate file 
df.to_csv('ordins_nums_dat.csv')
```

### Dummy Variables for Nominal Features 

#### Nominal features are categorical variables that do not preserve an order within them so they can be vectorized into binary features via dummy variables. 

```python
dummies = pd.get_dummies(df[nom], drop_first = True)
dummies.head()
```

```python
df = pd.concat([df, dummies], axis = 1)
```

```python
# save the transformed ordinal, numeric and nominal data along with their original untransformed version all into a file
df.to_csv('ordins_nums_dums_with_originals.dat.csv')
```

```python
# check one more time to make sure 'set' is the only feature of type object in our dataframe
for i in df.columns:
     if df[i].dtypes == 'object':
        print(i)
```
